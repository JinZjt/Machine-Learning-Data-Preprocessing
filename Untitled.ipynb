# 环境配置

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# input data

```python
data = pd.read_csv('Segmentation.csv')
data.head()
```

```python
data1 = pd.read_csv('Data.csv')
data1.head()
```

# label encode

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
dataset1 = data1.iloc[:,0]
dataset1 = labelencoder.fit_transform(dataset1)
onehotencoder = OneHotEncoder(categorical_features = [0])
dataset1 = onehotencoder.fit_transform(dataset1.reshape(-1,1)).toarray()
```

```python
labelencoder.classes_
```

# standardization&normalization

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
dataset = data.iloc[:,4:7].values # 为方便演示，选取多列数据构成矩阵，因为这两个函数只能以矩阵为参数
scaler1 = StandardScaler()
scaler2 = MinMaxScaler(feature_range =(0,1))
dataset = scaler1.fit_transform(dataset)
dataset = scaler2.fit_transform(dataset)
plt.plot(dataset)
```

由上图可见，已经都标准化且归一化了

## skewness visualization

```python
dataset = data.iloc[:,6].values
plt.subplot(1,4,1)
plt.hist(dataset, bins = 15, color = "Blue")
plt.subplot(1,4,2)
plt.hist(dataset**2, bins = 15, color = "Blue")
plt.subplot(1,4,3)
plt.hist(dataset**0.5, bins = 15, color = "DarkBlue")
plt.subplot(1,4,4)
plt.hist(1/dataset, bins = 15, color = "DarkBlue")
```

看分布，都分布在中间是比较好的，左倾右倾都有缺点，inverse还是很有效果的

```python
from scipy import stats
dataset_t, lamda = stats.boxcox(dataset)
```

# 相关度分析

```python
corr = data.corr()
sns.heatmap(corr)
```

图中白色的点代表正相关性较强的点，黑色的点是负相关性强的点。通过画热图，能对各变量之间相关性有一个大概的印象，方便后续的处理。
